```python
duration_split = df['duration'].str.split(' ', n=1, expand=True)

# conversion robuste : les 'Unknown', '', etc. deviennent NaN
df['duration_value'] = pd.to_numeric(duration_split[0], errors='coerce')

df['duration_unit'] = duration_split[1].str.strip()
```

```python
import psycopg2


# 1. Connexion à PostgreSQL (conteneur Docker)
conn = psycopg2.connect(
    host="localhost",
    port=5432,
    dbname="netflixetldb",   # adapte si tu as mis un autre nom
    user="SYSTEM",          # ou SYSTEM selon TON docker
    password="Postgres.123"
)

cur = conn.cursor()

# 2. Création de la table (on la supprime si elle existe déjà)
sql_create = """
DROP TABLE IF EXISTS netflix_titles;

CREATE TABLE netflix_titles (
  show_id         INTEGER,
  type            VARCHAR(20),
  title           VARCHAR(300),
  director        VARCHAR(300),
  cast_list       VARCHAR(1000),
  country         VARCHAR(300),
  date_added      DATE,
  release_year    INTEGER,
  rating          VARCHAR(10),
  duration        VARCHAR(50),
  listed_in       VARCHAR(500),
  description     VARCHAR(2000),
  year_added      INTEGER,
  month_added     INTEGER,
  duration_value  INTEGER,
  duration_unit   VARCHAR(20)
);
"""

cur.execute(sql_create)

# 3. Préparer les lignes à insérer à partir de ton DataFrame df
# (on suppose que df est déjà nettoyé comme tu l’as fait avant)

rows = [
    (
        int(row.show_id),
        row.type,
        row.title,
        row.director,
        row.cast,          # pandas garde le nom 'cast', on l’insère dans cast_list
        row.country,
        row.date_added,    # datetime pandas accepté par psycopg2
        int(row.release_year),
        row.rating,
        row.duration,
        row.listed_in,
        row.description,
        int(row.year_added),
        int(row.month_added),
        int(row.duration_value),
        row.duration_unit
    )
    for _, row in df.iterrows()
]

# 4. Requête d’insertion
sql_insert = """
INSERT INTO netflix_titles (
  show_id, type, title, director, cast_list, country,
  date_added, release_year, rating, duration,
  listed_in, description, year_added, month_added,
  duration_value, duration_unit
) VALUES (
  %s, %s, %s, %s, %s, %s,
  %s, %s, %s, %s,
  %s, %s, %s, %s,
  %s, %s
);
"""

cur.executemany(sql_insert, rows)
conn.commit()

cur.close()
conn.close()

print("Table netflix_titles créée et données insérées avec succès 🎉")
```
